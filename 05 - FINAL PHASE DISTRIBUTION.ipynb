#### LIBRARY IMPORTS

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
from datetime import datetime
```

#### LOAD FILE

```python
df = pd.read_csv('smk_stats_v10.csv', sep = ';', encoding = 'iso-8859-1')
```

#### Data Preparation

```python
# Les colonnes 2 à 24 doivent être transformées en Datetime pour pouvoir prendre en compte les chronos.
# Le traitement ci-dessous:
df['MC1'] = pd.to_datetime(df['MC1'], format='%M:%S,%f')
df['DP1'] = pd.to_datetime(df['DP1'], format='%M:%S,%f')
df['GV1'] = pd.to_datetime(df['GV1'], format='%M:%S,%f')
df['BC1'] = pd.to_datetime(df['BC1'], format='%M:%S,%f')
df['MC2'] = pd.to_datetime(df['MC2'], format='%M:%S,%f')
df['CI1'] = pd.to_datetime(df['CI1'], format='%M:%S,%f')
df['GV2'] = pd.to_datetime(df['GV2'], format='%M:%S,%f')
df['DP2'] = pd.to_datetime(df['DP2'], format='%M:%S,%f')
df['BC2'] = pd.to_datetime(df['BC2'], format='%M:%S,%f')
df['MC3'] = pd.to_datetime(df['MC3'], format='%M:%S,%f')
df['KB1'] = pd.to_datetime(df['KB1'], format='%M:%S,%f')
df['CI2'] = pd.to_datetime(df['CI2'], format='%M:%S,%f')
df['VL1'] = pd.to_datetime(df['VL1'], format='%M:%S,%f')
df['BC3'] = pd.to_datetime(df['BC3'], format='%M:%S,%f')
df['MC4'] = pd.to_datetime(df['MC4'], format='%M:%S,%f')
df['DP3'] = pd.to_datetime(df['DP3'], format='%M:%S,%f')
df['KB2'] = pd.to_datetime(df['KB2'], format='%M:%S,%f')
df['GV3'] = pd.to_datetime(df['GV3'], format='%M:%S,%f')
df['VL2'] = pd.to_datetime(df['VL2'], format='%M:%S,%f')
df['RR'] = pd.to_datetime(df['RR'], format='%M:%S,%f')
df['Total Time'] = pd.to_datetime(df['Total Time'], format='%M:%S,%f')
```

```python
# Identifier tous les temps effectués en Phase Finale de 2012 à 2019.
# Il va falloir instancier une variable par course sinon les lignes en 09:59,99 seront supprimées
final_phase_global_mc1 = df[(df['Phase'] != 'Group Stage') & (df['MC1'] != '1900-01-01 00:09:59.990')]
final_phase_global_dp1 = df[(df['Phase'] != 'Group Stage') & (df['DP1'] != '1900-01-01 00:09:59.990')]
final_phase_global_gv1 = df[(df['Phase'] != 'Group Stage') & (df['GV1'] != '1900-01-01 00:09:59.990')]
final_phase_global_bc1 = df[(df['Phase'] != 'Group Stage') & (df['BC1'] != '1900-01-01 00:09:59.990')]
final_phase_global_mc2 = df[(df['Phase'] != 'Group Stage') & (df['MC2'] != '1900-01-01 00:09:59.990')]
final_phase_global_ci1 = df[(df['Phase'] != 'Group Stage') & (df['CI1'] != '1900-01-01 00:09:59.990')]
final_phase_global_gv2 = df[(df['Phase'] != 'Group Stage') & (df['GV2'] != '1900-01-01 00:09:59.990')]
final_phase_global_dp2 = df[(df['Phase'] != 'Group Stage') & (df['DP2'] != '1900-01-01 00:09:59.990')]
final_phase_global_bc2 = df[(df['Phase'] != 'Group Stage') & (df['BC2'] != '1900-01-01 00:09:59.990')]
final_phase_global_mc3 = df[(df['Phase'] != 'Group Stage') & (df['MC3'] != '1900-01-01 00:09:59.990')]
final_phase_global_kb1 = df[(df['Phase'] != 'Group Stage') & (df['KB1'] != '1900-01-01 00:09:59.990')]
final_phase_global_ci2 = df[(df['Phase'] != 'Group Stage') & (df['CI2'] != '1900-01-01 00:09:59.990')]
final_phase_global_vl1 = df[(df['Phase'] != 'Group Stage') & (df['VL1'] != '1900-01-01 00:09:59.990')]
final_phase_global_bc3 = df[(df['Phase'] != 'Group Stage') & (df['BC3'] != '1900-01-01 00:09:59.990')]
final_phase_global_mc4 = df[(df['Phase'] != 'Group Stage') & (df['MC4'] != '1900-01-01 00:09:59.990')]
final_phase_global_dp3 = df[(df['Phase'] != 'Group Stage') & (df['DP3'] != '1900-01-01 00:09:59.990')]
final_phase_global_kb2 = df[(df['Phase'] != 'Group Stage') & (df['KB2'] != '1900-01-01 00:09:59.990')]
final_phase_global_gv3 = df[(df['Phase'] != 'Group Stage') & (df['GV3'] != '1900-01-01 00:09:59.990')]
final_phase_global_vl2 = df[(df['Phase'] != 'Group Stage') & (df['VL2'] != '1900-01-01 00:09:59.990')]
final_phase_global_rr = df[(df['Phase'] != 'Group Stage') & (df['RR'] != '1900-01-01 00:09:59.990')]
```

### Distribution des temps réalisé lors des différentes phases finales du championnat du monde de Time Trial entre 2012 et 2019.

```python
plt.figure(figsize = (20,10))

plt.subplot(2,2,1)
plt.hist(final_phase_global_mc1['MC1'],bins = 8, facecolor='black', ec='black', alpha = 0.5)
plt.title('MC1 (Final Phase - Global Distribution)', fontweight='bold')

plt.subplot(2,2,2)
plt.hist(final_phase_global_dp1['DP1'], bins = 8, facecolor='green',ec='black', alpha = 0.6)
plt.title('DP1 (Final Phase - Global Distribution)', fontweight='bold')

plt.subplot(2,2,3)
plt.hist(final_phase_global_gv1['GV1'], bins = 8, facecolor='brown',ec='black', alpha = 0.8)
plt.title('GV1 (Final Phase - Global Distribution)', fontweight='bold')

plt.subplot(2,2,4)
plt.hist(final_phase_global_bc1['BC1'], bins = 8, facecolor='red',ec='black', alpha = 0.7)
plt.title('BC1 (Final Phase - Global Distribution)', fontweight='bold')

plt.show()

plt.figure(figsize = (20,10))

plt.subplot(2,2,1)
plt.hist(final_phase_global_mc2['MC2'], bins = 8, facecolor='black',ec='black', alpha=0.5)
plt.title('MC2 (Final Phase - Global Distribution)', fontweight='bold')

plt.subplot(2,2,2)
plt.hist(final_phase_global_ci1['CI1'], bins = 8, facecolor='orange',ec='black', alpha = 0.5)
plt.title('CI1 (Final Phase - Global Distribution)', fontweight='bold')

plt.subplot(2,2,3)
plt.hist(final_phase_global_gv2['GV2'], bins = 8, facecolor='brown',ec='black', alpha=0.8)
plt.title('GV2 (Final Phase - Global Distribution)', fontweight='bold')

plt.subplot(2,2,4)
plt.hist(final_phase_global_dp2['DP2'], bins = 8, facecolor='green',ec='black', alpha=0.6)
plt.title('DP2 (Final Phase - Global Distribution)', fontweight='bold')

plt.show()

plt.figure(figsize = (20,10))

plt.subplot(2,2,1)
plt.hist(final_phase_global_bc2['BC2'], bins = 8, facecolor='red',ec='black', alpha=0.7)
plt.title('BC2 (Final Phase - Global Distribution)', fontweight='bold')

plt.subplot(2,2,2)
plt.hist(final_phase_global_mc3['MC3'], bins = 8, facecolor='black',ec='black', alpha=0.5)
plt.title('MC3 (Final Phase - Global Distribution)', fontweight='bold')

plt.subplot(2,2,3)
plt.hist(final_phase_global_kb1['KB1'], bins = 8, facecolor='blue',ec='black', alpha=0.5)
plt.title('KB1 (Final Phase - Global Distribution)', fontweight='bold')

plt.subplot(2,2,4)
plt.hist(final_phase_global_ci2['CI2'], bins = 8, facecolor='orange',ec='black', alpha=0.5)
plt.title('CI2 (Final Phase - Global Distribution)', fontweight='bold')

plt.show()

plt.show()

plt.figure(figsize = (20,10))

plt.subplot(2,2,1)
plt.hist(final_phase_global_vl1['VL1'], bins = 8, facecolor='cyan',ec='black', alpha=0.5)# 137 bins réels
plt.title('VL1 (Final Phase - Global Distribution)', fontweight='bold')

plt.subplot(2,2,2)
plt.hist(final_phase_global_bc3['BC3'], bins = 8, facecolor='red',ec='black', alpha=0.7)
plt.title('BC3 (Final Phase - Global Distribution)', fontweight='bold')

plt.subplot(2,2,3)
plt.hist(final_phase_global_mc4['MC4'], bins = 8, facecolor='black',ec='black', alpha=0.5)
plt.title('MC4 (Final Phase - Global Distribution)', fontweight='bold')

plt.subplot(2,2,4)
plt.hist(final_phase_global_dp3['DP3'], bins = 8, facecolor='green',ec='black', alpha=0.6)
plt.title('DP3 (Final Phase - Global Distribution)', fontweight='bold')

plt.show()

plt.figure(figsize = (20,10))

plt.subplot(2,2,1)
plt.hist(final_phase_global_kb2['KB2'], bins = 8, facecolor='blue',ec='black', alpha=0.5)
plt.title('KB2 (Final Phase - Global Distribution)', fontweight='bold')

plt.subplot(2,2,2)
plt.hist(final_phase_global_gv3['GV3'], bins = 8, facecolor='brown',ec='black', alpha=0.8)
plt.title('GV3 (Final Phase - Global Distribution)', fontweight='bold')

plt.subplot(2,2,3)
plt.hist(final_phase_global_vl2['VL2'], bins = 8, facecolor='cyan',ec='black', alpha=0.5)
plt.title('VL2 (Final Phase - Global Distribution)', fontweight='bold')

plt.subplot(2,2,4)
plt.hist(final_phase_global_rr['RR'], bins = 8, facecolor='purple',ec='black', alpha=0.6)
plt.title('RR (Final Phase - Global Distribution)', fontweight='bold')

plt.show()
```
