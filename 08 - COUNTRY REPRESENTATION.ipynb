### 7) -	Représentation des joueurs par pays de 2012 à 2019

#### A) Représentation globale des participants

```python
# Il faut instancier une variable par année de 2012 à 2019 pour le top 16 de chaque année en phase de poule.
tt_2012 = df[(df['Phase'] == 'Group Stage') & (df['Year'] == 2012)]
tt_2013 = df[(df['Phase'] == 'Group Stage') & (df['Year'] == 2013)]
tt_2014 = df[(df['Phase'] == 'Group Stage') & (df['Year'] == 2014)]
tt_2015 = df[(df['Phase'] == 'Group Stage') & (df['Year'] == 2015)]
tt_2016 = df[(df['Phase'] == 'Group Stage') & (df['Year'] == 2016)]
tt_2017 = df[(df['Phase'] == 'Group Stage') & (df['Year'] == 2017)]
tt_2018 = df[(df['Phase'] == 'Group Stage') & (df['Year'] == 2018)]
tt_2019 = df[(df['Phase'] == 'Group Stage') & (df['Year'] == 2019)]
```

```python
from matplotlib.ticker import MaxNLocator
plt.figure(figsize = (14,7))
#ax = plt.figure().gca()
#ax.yaxis.set_major_locator(MaxNLocator(integer=True))
plt.subplot(2,2,1)
plt.hist(tt_2019['Country'],bins = 11, ec='black',alpha = 0.5)
plt.title('2019 - Country Distribution', fontweight='bold')


plt.subplot(2,2,2)
plt.hist(tt_2018['Country'],bins = 9, ec='black', alpha = 0.5)
plt.title('2018 - Country Distribution', fontweight='bold')

plt.subplot(2,2,3)
plt.hist(tt_2017['Country'],bins = 9, ec='black', alpha = 0.5)
plt.title('2017 - Country Distribution', fontweight='bold')

plt.subplot(2,2,4)
plt.hist(tt_2016['Country'],bins = 8, ec='black', alpha = 0.5)
plt.title('2016 - Country Distribution', fontweight='bold')

plt.show()

plt.figure(figsize = (14,7))

plt.subplot(2,2,1)
plt.hist(tt_2015['Country'],bins = 9, ec='black',alpha=0.5)
plt.title('2015 - Country Distribution', fontweight='bold')

plt.subplot(2,2,2)
plt.hist(tt_2014['Country'],bins = 7, ec='black',alpha = 0.5)
plt.title('2014 - Country Distribution', fontweight='bold')

plt.subplot(2,2,3)
plt.hist(tt_2013['Country'],bins = 6, ec='black',alpha=0.5)
plt.title('2013 - Country Distribution', fontweight='bold')

plt.subplot(2,2,4)
plt.hist(tt_2012['Country'],bins = 8, ec='black',alpha=0.5)
plt.title('2012 - Country Distribution', fontweight='bold')

plt.show()#11,9,9,8,9,7,6,8
```

#### Commentaires:
- La France est le pays qui compte le plus de participants, peu importe l'année.
- Le championnat du monde est particulièrement à consonnance européenne (France, Grande-Bretagne, Pays-bas représentent à eux seuls les 3/4 des participants).
- On constate une forte chute des participants français en 2019 (toujours majoritaires, mais de peu devant la Grande-Bretagne et la Hollande).
- 2019 est l'édition avec le plus de nationalités différentes (11 au total).
- 2019 est également l'édition avec le plus de participants venant de continents différents (5 continents représentés au total).

#### B) Représentation des participants au top 16 par pays

```python
# Il faut instancier une variable par année de 2012 à 2019 pour le top 16 de chaque année en phase de poule.
top16_2012 = df[(df['Phase'] == 'Top 16') & (df['Year'] == 2012) & (df['Position'] < 17)]
top16_2013 = df[(df['Phase'] == 'Top 16') & (df['Year'] == 2013) & (df['Position'] < 17)]
top16_2014 = df[(df['Phase'] == 'Top 16') & (df['Year'] == 2014) & (df['Position'] < 17)]
top16_2015 = df[(df['Phase'] == 'Top 16') & (df['Year'] == 2015) & (df['Position'] < 17)]
top16_2016 = df[(df['Phase'] == 'Top 16') & (df['Year'] == 2016) & (df['Position'] < 17)]
top16_2017 = df[(df['Phase'] == 'Top 16') & (df['Year'] == 2017) & (df['Position'] < 17)]
top16_2018 = df[(df['Phase'] == 'Top 16') & (df['Year'] == 2018) & (df['Position'] < 17)]
top16_2019 = df[(df['Phase'] == 'Last Elim Out') & (df['Year'] == 2019) & (df['Position'] < 17)]
```

```python
from matplotlib.ticker import MaxNLocator
plt.figure(figsize = (14,7))

plt.subplot(2,2,1)
plt.hist(top16_2019['Country'],bins = 7, ec='black',alpha = 0.5)
plt.title('2019 - Country Distribution in Top 16', fontweight='bold')


plt.subplot(2,2,2)
plt.hist(top16_2018['Country'],bins = 6, ec='black', alpha = 0.5)
plt.title('2018 - Country Distribution in Top 16', fontweight='bold')

plt.subplot(2,2,3)
plt.hist(top16_2017['Country'],bins = 6, ec='black', alpha = 0.5)
plt.title('2017 - Country Distribution in Top 16', fontweight='bold')

plt.subplot(2,2,4)
plt.hist(top16_2016['Country'],bins = 5, ec='black', alpha = 0.5)
plt.title('2016 - Country Distribution in Top 16', fontweight='bold')

plt.show()

plt.figure(figsize = (14,7))

plt.subplot(2,2,1)
plt.hist(top16_2015['Country'],bins = 4, ec='black',alpha=0.5)
plt.title('2015 - Country Distribution in Top 16', fontweight='bold')

plt.subplot(2,2,2)
plt.hist(top16_2014['Country'],bins = 4, ec='black',alpha = 0.5)
plt.title('2014 - Country Distribution in Top 16', fontweight='bold')

plt.subplot(2,2,3)
plt.hist(top16_2013['Country'],bins = 4, ec='black',alpha=0.5)
plt.title('2013 - Country Distribution in Top 16', fontweight='bold')

plt.subplot(2,2,4)
plt.hist(top16_2012['Country'],bins = 5, ec='black',alpha=0.5)
plt.title('2012 - Country Distribution in Top 16', fontweight='bold')

plt.show()
```

#### Commentaires:
- Les français, participants les plus nombreux chaque année, sont aussi la nation la plus représentée en phase finale de Time Trial sur la période analysée.
- Le nombre de français qualifiés pour le top 16 est en légère diminution depuis 2018.
- Il y a 7 nations différentes représentées dans le top 16 en 2019, un record.
