#### LIBRARY IMPORTS

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
from datetime import datetime
import time
```

#### LOAD FILE

```python
df = pd.read_csv('smk_stats_v10.csv', sep = ';', encoding = 'iso-8859-1')
```

#### Data preparation

```python
# TRANSFORMATION DES COLONNES 2 à 24 en Datetime pour pouvoir prendre en compte les chronos.
df['MC1'] = pd.to_datetime(df['MC1'], format='%M:%S,%f')
df['DP1'] = pd.to_datetime(df['DP1'], format='%M:%S,%f')
df['GV1'] = pd.to_datetime(df['GV1'], format='%M:%S,%f')
df['BC1'] = pd.to_datetime(df['BC1'], format='%M:%S,%f')
df['MC2'] = pd.to_datetime(df['MC2'], format='%M:%S,%f')
df['CI1'] = pd.to_datetime(df['CI1'], format='%M:%S,%f')
df['GV2'] = pd.to_datetime(df['GV2'], format='%M:%S,%f')
df['DP2'] = pd.to_datetime(df['DP2'], format='%M:%S,%f')
df['BC2'] = pd.to_datetime(df['BC2'], format='%M:%S,%f')
df['MC3'] = pd.to_datetime(df['MC3'], format='%M:%S,%f')
df['KB1'] = pd.to_datetime(df['KB1'], format='%M:%S,%f')
df['CI2'] = pd.to_datetime(df['CI2'], format='%M:%S,%f')
df['VL1'] = pd.to_datetime(df['VL1'], format='%M:%S,%f')
df['BC3'] = pd.to_datetime(df['BC3'], format='%M:%S,%f')
df['MC4'] = pd.to_datetime(df['MC4'], format='%M:%S,%f')
df['DP3'] = pd.to_datetime(df['DP3'], format='%M:%S,%f')
df['KB2'] = pd.to_datetime(df['KB2'], format='%M:%S,%f')
df['GV3'] = pd.to_datetime(df['GV3'], format='%M:%S,%f')
df['VL2'] = pd.to_datetime(df['VL2'], format='%M:%S,%f')
df['RR'] = pd.to_datetime(df['RR'], format='%M:%S,%f')
df['Total Time'] = pd.to_datetime(df['Total Time'], format='%M:%S,%f')
```

```python
# Identifier et instancier de manière globale le Top 8 en Groupe Stage sur l'ensemble des années
top8_group_stage = df[(df['Phase'] == 'Group Stage') & (df['Position'] < 9)]
```

```python
# Il faut instancier une variable par année de 2012 à 2019 pour le top 8 de chaque année en phase de poule.
top8_2012_group_stage = df[(df['Phase'] == 'Group Stage') & (df['Year'] == 2012) & (df['Position'] < 9)]
top8_2013_group_stage = df[(df['Phase'] == 'Group Stage') & (df['Year'] == 2013) & (df['Position'] < 9)]
top8_2014_group_stage = df[(df['Phase'] == 'Group Stage') & (df['Year'] == 2014) & (df['Position'] < 9)]
top8_2015_group_stage = df[(df['Phase'] == 'Group Stage') & (df['Year'] == 2015) & (df['Position'] < 9)]
top8_2016_group_stage = df[(df['Phase'] == 'Group Stage') & (df['Year'] == 2016) & (df['Position'] < 9)]
top8_2017_group_stage = df[(df['Phase'] == 'Group Stage') & (df['Year'] == 2017) & (df['Position'] < 9)]
top8_2018_group_stage = df[(df['Phase'] == 'Group Stage') & (df['Year'] == 2018) & (df['Position'] < 9)]
top8_2019_group_stage = df[(df['Phase'] == 'Group Stage') & (df['Year'] == 2019) & (df['Position'] < 9)]
```
