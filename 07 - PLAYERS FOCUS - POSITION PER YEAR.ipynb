### Focus analytique sur les position des joueurs en phase de poule de Time Trial de 2012 à 2019

#### A) TOP 8

#### Data Prep

```python
# Identifier de manière globale le Top 8 en Groupe Stage sur l'ensemble des années
top8_group_stage = df[(df['Phase'] == 'Group Stage') & (df['Position'] < 9)]
```

```python
# Définir les variables x et y pour le lineplot par année
x_top8_gstage = top8_group_stage['Year']
y_top8_gstage = top8_group_stage['Position']
```

```python
# Create list with personnalized colors -> will be used for graphs
colors = ['#0e469a','#9c1717','#0a0eff','#626360','#00a86b','#db7027','#004620','#000000','#f000ff','#fe2e2e','#f2ae42',
          '#98bf2a','#8b4306','#613dbd','#ffd300','#00ffff']
```

```python
# Lineplot Seaborn pour visualiser les années et les positions
plt.figure(figsize = (20,10))

sns.set(style="darkgrid")
sns.set_palette(sns.color_palette(colors))# Récupérer la palette de couleur personnalisée ici
sns.lineplot( x_top8_gstage, y_top8_gstage, hue="Nickname",marker="o", data = top8_group_stage)

plt.legend(loc='upper left', bbox_to_anchor=(1, 1), ncol=1, fontsize='20')# Définition de la légende en dehors du graphique

plt.xlabel('Year', fontweight='bold', fontsize = 16)
plt.ylabel('Position', fontweight='bold', fontsize = 16)
plt.title('Représentation des joueurs présents dans le top 8 de Phase de poule de Time Trial entre 2012 et 2019', fontweight='bold', 
          fontsize = 20)

# Inverser l'axe des y
ax = plt.gca()
ax.set_ylim(ax.get_ylim()[::-1])

plt.show()
```

#### Commentaires au sujet du graphe:
- 16 personnes différentes ont réussi à se positionner dans le top 8 dans la phase de Group Stage entre 2012 et 2019.
- 6 joueurs ont réussi à se positionner une seule fois dans le top 8 (Chraizy,Lafungo,JDR, MF, Jey et Monkey Kong).
- Godfather of SMK est impressionnant de régularité (Sur 8 années, il réalisé l'exploit d'être 7 fois sur 8 1er de la poule de Time Trial, sauf en 2017). 5 top 1 consécutifs entre 2012 et 2016.
- Neo est lui aussi très régulier et performant (toujours dans le top 3 entre 2012 et 2018)
- ScouB est de plus en plus performant en Time Trial depuis 2016 (toujours dans le top 4 en poule entre 2016 et 2019).
- KVD est toujours très bien positionné (toujours dans le top 4 7 fois sur 8, 3 fois de suite Top 2 de 2017 à 2019).
- Antistar toujours bien placé également (entre la 3ème et 5ème place de 2013 à 2019).
- Gatchan et Kedjyar sont également présent de manière fréquente entre le top 4 et le top 7 lors de leurs participations.
- Masterpie9 est également présents de manière régulière entre le top 6 et le top 8, sauf sur les deux dernières années 
analysées (2018 et 2019).
- Geo a accroché la 8ème place 3 fois en 2013,2015 et 2016.
- Championne SMK est apparue dans le top 8 depuis 2017 en arrivant à se positionner entre la 6ème et 8ème place 3 fois de suite.
- 4 nouveaux joueurs ont intégré le top 8 lors des deux dernières éditions, ces joueurs étant placés entre la 6ème et 9ème place (JDR, MF, Lafungo et Chraizy) 
- Seulement 6 joueurs différents se sont placés dans le top 4 des phases de poule de Time Trial de 2012 à 2019 (Sami, Neo, KVD, Gatchan, Antistar, Kedjyar et ScouB).

#### Conclusion sur le graphe:
- Accéder au top 8 lors de la phase de poule reste réservé à une élite. En effet, seul 10 joueurs différents ont réussi à se placer 2 fois ou plus dans le top 8 ces 8 dernières années.

#### D) Positions des joueurs: match pour la 3ème place (2012 à 2019)

```python
# Compliqué à gérer car il faut gérer les données de 2019 dont les données de 3ème et 4ème place sont relatives à la
# position finale, et non pas à la position lors de la phase en question.
# Comment gérer? 
# 1) Extraire les données phase = Match for 3rd place
# 2) Extraire les données Last Elim Out avec une position égale à 3 ou à 4.
# 3) Regrouper les deux dataframes
```

```python
# 1) Extraire les données phase = Match for 3rd place
third_p1 = df[(df['Phase'] == 'Match for 3rd Place')]
```

```python
# 2) Extraire les données Last Elim Out avec une position égale à 3 ou à 4.
third_p2 = df[(df['Phase'] == 'Last Elim Out') & ((df['Position'] == 3) |(df['Position'] == 4))]
```

```python
# 3) Update la value des deux lignes présentes dans third_p2 avec 1 à la place de 3 et 2 à la place de 4
third_p2.at[40,'Position']= 2 # On récupère l'index 40
third_p2.at[42,'Position']= 1 # On récupère l'index 42
third_p2.head()
```

```python
# 4) Regrouper les deux variables dans un dataframe.
frames = [third_p1,third_p2]# les deux variables sont stockées dans une liste
third_place = pd.concat(frames) # Puis on utilise le concat sur la liste créée plus haut.
```

```python
# Définition des variables d'abcisse et ordonnée pour le lineplot des matchs pour la 3ème place
x_3rd_place = third_place['Year']
y_3rd_place = third_place['Position']
```

```python
# Palette de couleurs personnalisée
colors_3rd_place = ['#626360','#fe2e2e','#00a86b','#f576a3','#0a0eff','#db7027','#000000','#98bf2a','#9c1717']
```

```python
import math
y =[1,2] # On définit une liste qui va nous servir à définir l'axe des y pour nos histogrammes du top 8.
yint = range(min(y), math.ceil(max(y))+1) # Permet d'incrémenter l'axe y de 1 à 2
```

```python
# Lineplot Seaborn pour visualiser les années et les positions (Match for 3rd place)
plt.figure(figsize = (20,10))

sns.set(style="darkgrid")
sns.set_palette(sns.color_palette(colors_3rd_place))# Récupérer la palette de couleur personnalisée ici
sns.lineplot( x_3rd_place, y_3rd_place, hue="Nickname",marker="o", data = third_place)
plt.yticks(yint)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1), ncol=1, fontsize='19')# Définition de la légende en dehors du graphique

plt.xlabel('Year', fontweight='bold', fontsize = 16)
plt.ylabel('Position', fontweight='bold', fontsize = 16)
plt.title('Positions des joueurs suite aux matchs pour la 3ème place disputés entre 2012 et 2019 en Time Trial', fontweight='bold', 
          fontsize = 20)
plt.gca().invert_yaxis()
plt.show()
```

#### Commentaires:
